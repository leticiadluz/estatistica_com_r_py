# 1 - Variveis Aleatrias Discretas

- Uma varivel aleatria pode ser entendida como uma varivel quantitativa, cujo resultado (valor) depende de fatores aleatrios.  Os valores que ela pode assumir esto sujeitos  incerteza e no podem ser predeterminados com certeza antes da ocorrncia do evento ou experimento. A quantificao dessa incerteza  feita atravs das probabilidades associadas a cada possvel valor da varivel aleatria.

### Variveis aleatrias discretas so aquelas que assumem valores em um conjunto finito ou infinito enumervel de pontos. Isso significa que os valores que a varivel pode assumir so distintos e podem ser contados individualmente, mas no podem ser medidos em uma escala contnua. Aqui esto alguns exemplos de variveis aleatrias discretas:
- Nmero de filhos em uma famlia: Esta varivel aleatria conta o nmero de filhos em cada famlia. Os valores possveis so nmeros inteiros no negativos, como 0, 1, 2, 3, e assim por diante.
- Nmero de caras ao lanar uma moeda vrias vezes: Se voc lanar uma moeda vrias vezes, a varivel aleatria aqui  o nmero de vezes que ela mostra "cara". Os valores possveis so 0 (nenhuma cara), 1 (uma cara), 2 (duas caras) e assim por diante.
- Nmero de alunos que passam em um exame: Suponha que h um exame com uma taxa de aprovao de 70%. A varivel aleatria aqui  o nmero de alunos que passam no exame. Os valores possveis so 0 (nenhum aluno passa), 1 (um aluno passa), 2 (dois alunos passam) e assim por diante.
- Nmero de acidentes em um cruzamento em um determinado dia: Esta varivel aleatria representa o nmero de acidentes ocorridos em um cruzamento durante um dia. Os valores possveis so nmeros inteiros no negativos

## 1.1 Valor mdio de uma varivel aleatria discreta
- O valor esperado, tambm conhecido como mdia ponderada ou expectativa matemtica, ou ainda esperana matemtica  um conceito fundamental na teoria da probabilidade e na estatstica. Ele representa uma medida de centro de uma distribuio de probabilidade e fornece uma estimativa do valor mdio de uma varivel aleatria ao longo de um grande nmero de repeties de um experimento aleatrio.
- Matematicamente, o valor esperado de uma varivel aleatria X, denotado como E[X] ou ,  calculado como a soma do produto dos possveis valores que a varivel aleatria pode assumir e suas respectivas probabilidades de ocorrncia. Para uma varivel aleatria discreta X, o valor esperado  dado por:



- Xi: so os possveis valores que a varivel aleatria X pode assumir.
- P(X=xi):  a probabilidade de que a varivel aleatria X assuma o valor xi.

**Por exemplo, se estivermos jogando um dado justo de seis faces, o valor esperado do resultado do lanamento  calculado como:**
- E[X] =  (1 X 1/6) + (2 X 1/6) + (3 X 1/6) + (4 X 1/6) + (5 X 1/6) + (6 X 1/6) = 3.5

- Isso significa que, em mdia, esperaramos obter um resultado de aproximadamente 3.5 aps um grande nmero de lanamentos do dado justo.
- Portanto, o valor esperado representa o valor mdio que esperaramos observar ao longo de mltiplas repeties do experimento aleatrio. Em outras palavras, ele representa o valor mdio que esperamos para a varivel aleatria, considerando todas as possveis ocorrncias e suas respectivas probabilidades.

## 1.2 Varincia de uma varivel aleatria discreta
- A varincia de uma varivel aleatria discreta  uma medida de disperso que indica o quo longe os valores da varivel esto distribudos em relao  mdia. A  a frmula da varincia Var(X) de uma varivel aleatria discreta X pode ser expressa como:



- xi: so os valores possveis da varivel aleatria
- E(X):  o valor esperado (mdia) da varivel aleatria X
- pi: so as probabilidades associadas a cada valor xi
- n:  o nmero total de valores possveis da varivel aleatria X

**Exemplo:  Um empresrio tem um lucro de 100 reais por pea feita sem defeito e de 30 reais por pea feita com defeito. Existe uma probabilidade de 20% de que a pea saia com defeito. Qual lucro esperado? E qual a varincia do valor esperado?**
- E(X)=(1000.8)+(300.2)=80+6= 86 reais de lucro esperado
- Varincia:
    - Para x1 = 100(pea sem defeito):
        - (x1E(X)).p1= (10086) 0.8 = 196  0.8 = 156.8 
    - Para x2 = 30(pea com defeito):
        - (x2E(X)).p2= (3086) 0.2 = 3136 x 0.2 = 627.2

## 2.4 Distribuio de Bernoulli

## 2.5 Distribuio Binomial
- **A distribuio binomial  uma distribuio de probabilidade discreta que descreve o nmero de sucessos em uma sequncia fixa de tentativas independentes**, onde cada tentativa tem apenas dois resultados possveis: sucesso ou fracasso. Ela  chamada de **binomial** porque envolve duas possibilidades, sucesso ou fracasso.

- Os principais parmetros da distribuio binomial so:
    - n: o nmero total de tentativas ou ensaios.
    - p: a probabilidade de sucesso em cada tentativa.
- Chama-se de experimento binomial ao experimento:
    - que consiste em n ensaios de Bernoulli
    - cujos ensaios so independentes
    - para o qual a probabilidade de sucesso em cada ensaio  sempre igual a p, 0 < p < 1.


- A distribuio binomial  usada para modelar situaes em que h apenas dois resultados possveis em cada tentativa, como lanamentos de moedas, ensaios clnicos, teste de produtos, entre outros.  

### 2.5.1 Funo de Massa de Probabilidade (FP) da Distribuio Binomial:



- P(X=k) a probabilidade de obter exatamente k sucessos em n tentativas,
- (n/k) o coeficiente binomial, que representa o nmero de maneiras de escolher k sucessos em n tentativas e  calculado como:
    - n!/k!(n-k)!
- P  a probabilidade de sucesso em uma tentativa individual,
- (1p)  a probabilidade de fracasso em uma tentativa individual.
- k  o nmero de sucessos desejados.
---
- **Exemplo:** Calcule a probabilidade de obter exatamente duas caras em trs lanamentos de uma moeda:
    - n: nmero de tentativas (lanamentos) = 3
    - k: nmero de sucessos desejados (caras) = 2
    - p: probabilidade de sucesso em cada tentativa (cara) = 0.5 (uma moeda justa)
    - P(x=2) = (n/k) = 3/2, fatorial:
        - 3!/2!(3-2)!
        - 3!/2!1!
        - 3
   - P(x=2)= 3 x (0.5) X (1-0.5)^3-2
   - P(x=2)= 3 x (0.5) x (0.5)
   - P(x=2)= 3 x 0.25 x 0.5
   - P(X=2)= 0.375

### 2.5.2 Funo de Distribuio Acumulada (FDA) da Distribuio Binomial:



- P(Xk)  a probabilidade acumulada de obter k ou menos sucessos
- A soma  feita desde i=0 at i=k.
- (n/i)  o coeficiente binomial, que conta o nmero de maneiras de escolher i sucessos em n tentativas.
- p  a probabilidade de ter i sucessos
- (1p)  a probabilidade de fracasso em uma tentativa individual.
---
- **Exemplo:** Imagine um evento aleatrio dado pelo lanamento de uma moeda honesta e homognea. Se esta moeda for lanada 4 vezes, qual  a probabilidade de sair pelo menos 3 caras?

### 2.6.1 Funo de Massa de Probabilidade (FP) da Distribuio Hipergeomtrica:

---

- n = 5: nmero de itens na amostra.
- k = 0: nmero de itens defeituosos na amostra.
- Clculo dos Coeficientes Binomiais, n!/k!(n-k)!
    - (10/0) = 1
    - (90/5) = 43.949.268
    - (100/5) = 75.287.520
- P(x=0) = (1 x 43.949.268) / 75.287.520
- P(x=0) = 0.584

**Utilizamos a funo dhyper**

```r
# Sucessos desejados, sucessos da populao,  tamanho da populao - sucessos na populao, tamanho da amostra
dhyper(0, 10, 100-10, 5)
```
